The dataset consists of 100,000 records and 28 columns, capturing a wide range of financial and behavioral information about individual customers. Columns like ID, Customer_ID, Name, and SSN appear to be identifiers and may not contribute meaningfully to prediction tasks. ### Variables Description

The dataset comprises 28 variables that provide comprehensive information about customers, including demographic, financial, and behavioral attributes. Each record has an `ID` and a corresponding `Customer_ID`, along with a `Month` column representing time progression (from 1 to 8) for temporal analysis. Personal details include the customer’s `Name`, `Age`, and `SSN` (Social Security Number). Financial attributes capture the `Occupation`, `Annual_Income`, `Monthly_Inhand_Salary`, `Num_Bank_Accounts`, `Num_Credit_Card`, `Interest_Rate`, and `Num_of_Loan`. The `Type_of_Loan` field lists different types of loans, while `Delay_from_due_date` and `Num_of_Delayed_Payment` indicate the extent and frequency of payment delays. The `Changed_Credit_Limit` shows modifications in credit limits, and `Num_Credit_Inquiries` reflects how often credit reports were accessed. The `Credit_Mix` denotes the variety and quality of credit types used. Other financial metrics include `Outstanding_Debt`, `Credit_Utilization_Ratio`, `Credit_History_Age`, `Total_EMI_per_month`, `Amount_invested_monthly`, and `Monthly_Balance`. Behavioral indicators such as `Payment_of_Min_Amount` and `Payment_Behaviour` provide insight into payment habits. The target variable is `Credit_Score`, which categorizes customers as having a Good, Standard, or Poor credit rating.

#### Chart - 9

````python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Credit_Score', y='Outstanding_Debt', data=data_required, palette='Set2')
plt.title('Outstanding Debt by Credit Score')
plt.xlabel('Credit Score')
plt.ylabel('Outstanding Debt')
plt.grid(True)
plt.tight_layout()
plt.show()
````

##### 1. Why did you pick the specific chart?

We chose a **boxplot** to visualize the distribution of `Outstanding_Debt` across different `Credit_Score` categories (`Good`, `Standard`, and `Poor`). Boxplots are ideal for highlighting **medians, variability, and outliers**, which is critical in financial data.


##### 2. What is/are the insight(s) found from the chart?

 Individuals with **Poor credit scores** tend to have **higher median debt** and greater spread, indicating potential financial instability.
- Those with **Good credit scores** usually maintain **lower debt levels**, and their debt is more tightly distributed.
- The **Standard credit score** group lies between the two in terms of both median and spread.
- Significant **outliers** exist in all categories, but are particularly visible in the Good and Standard groups, possibly representing individuals managing large debts responsibly.
